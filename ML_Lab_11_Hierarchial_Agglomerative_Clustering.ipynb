```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv('student.csv')
print(df)

x = df[['cgpa','iq']].values #predictor variables

#to find number of clusters
dendro = sch.dendrogram(sch.linkage(x, method="single"))  #applying single (min) linkage
plt.title("Dendrogram Plot")
plt.ylabel("Euclidean Distances")
plt.xlabel("Students")
plt.show()

#from above dendrogram we can see 4 prominent/big clusters, so n_clusters=5
hc= AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='single')
y_predict= hc.fit_predict(x) #fit model

#scatter clusters
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1],label = 'Cluster 1') #for cluster 1
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1],label = 'Cluster 2') #for cluster 2
plt.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1],label = 'Cluster 3') #for cluster 3
plt.scatter(x[y_predict== 3, 0], x[y_predict == 3, 1],label = 'Cluster 4') #for cluster 4
plt.title('Clusters of Students')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.legend()
plt.show()
```
