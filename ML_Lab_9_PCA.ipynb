```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df=pd.read_csv("Acoustic Features.csv")
print(df)

#standardizing input features
x=df.drop(['Class'],axis=1)
st_x= StandardScaler()
x= st_x.fit_transform(x)

#target class
le = LabelEncoder()
y=le.fit_transform(df['Class'])

#dimensionality reduction
pca = PCA(n_components=3) #n_components = no. of principal components after dimensionality reduction
pca_features = pca.fit_transform(x)

#reduced dataset
pca_df = pd.DataFrame(data=pca_features, columns=['PC1', 'PC2', 'PC3'])

pca_df['Class'] = y
pca_df['Class'] = df['Class']
print(pca_df)

r,h,s,a= pca_df[pca_df['Class']=='relax'],pca_df[pca_df['Class']=='happy'],pca_df[pca_df['Class']=='sad'],pca_df[pca_df['Class']=='angry']
plt.scatter(r['PC1'], r['PC2'])
plt.scatter(h['PC1'], h['PC2'])
plt.scatter(s['PC1'], s['PC2'])
plt.scatter(a['PC1'], a['PC2'])
plt.legend(["relax", "happy", "sad","angry"])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

ax = plt.axes(projection ='3d')
ax.scatter(r['PC1'], r['PC2'], r['PC3'])
ax.scatter(h['PC1'], h['PC2'], h['PC3'])
ax.scatter(s['PC1'], s['PC2'], s['PC3'])
ax.scatter(a['PC1'], a['PC2'], a['PC3'])
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.legend(["relax", "happy", "sad","angry"])

```

```python
#target class
le = LabelEncoder()
y=le.fit_transform(df['Class'])
y
```
