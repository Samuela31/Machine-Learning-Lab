```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("Iris.csv")
print(df)

x=df[['SepalLengthCm','PetalWidthCm']].values #to get np array .values is used, else it'll show error in scattering
le = LabelEncoder()
y=le.fit_transform(df['Species'])

#since it is non-linear we use radial basis function kernel. Gamma value can be adjusted for adjusting boundary lines
model = svm.SVC(kernel="rbf", gamma=0.7)
model.fit(x, y)

#plotting decision boundary
DecisionBoundaryDisplay.from_estimator(model,x,response_method="predict") #can be used for linear svm also
plt.scatter(x[:, 0], x[:, 1], c=y,edgecolors='black') #edgecolors is used to outline points
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```
