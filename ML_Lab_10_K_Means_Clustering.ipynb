```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


df=pd.read_csv("student.csv")
print(df)

x = df[['cgpa','iq']].values #predictor variables

#We use Elbow graph method to find optimal number of clusters (K value)
#WCSS (Within-Cluster Sum of Square) is the sum of the squared distance between each point and the centroid in a cluster.
#When we plot the WCSS with the K value, the plot looks like an Elbow.
#As the number of clusters increases, the WCSS value will start to decrease.
wcss_list= []  #list of values of WCSS

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss_list)
plt.title('Elbow Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('WCSS list')
plt.show()

#from above graph we can see that elbow point is at 4
kmeans = KMeans(n_clusters=4)  #so no. of clusters=4 (k value)

#fitting model
y_predict= kmeans.fit_predict(x)

#scatter clusters
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1],label = 'Cluster 1') #for cluster 1
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1],label = 'Cluster 2') #for cluster 2
plt.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1],label = 'Cluster 3') #for cluster 3
plt.scatter(x[y_predict== 3, 0], x[y_predict == 3, 1],label = 'Cluster 4') #for cluster 4
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],marker = '^',label = 'Centroid') #show cluster centroid
plt.title('Clusters of Students')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


df=pd.read_csv("Iris.csv")
df
```

```python
x = df[['SepalLengthCm','PetalLengthCm']].values #predictor variables

#We use Elbow graph method to find optimal number of clusters (K value)
#WCSS is the sum of the squared distance between each point and the centroid in a cluster.
#When we plot the WCSS with the K value, the plot looks like an Elbow.
#As the number of clusters increases, the WCSS value will start to decrease.
wcss_list= []  #list of values of WCSS

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss_list)
plt.title('Elbow Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('WCSS list')
plt.show()
```

```python
#from above graph we can see that elbow point is at 3
kmeans = KMeans(n_clusters=3)  #so no. of clusters=3 (k value)

#fitting model
y_predict= kmeans.fit_predict(x)
```

```python
#scatter clusters
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1],label = 'Cluster 1') #for cluster 1
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1],label = 'Cluster 2') #for cluster 2
plt.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1],label = 'Cluster 3') #for cluster 3
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],marker = '^',label = 'Centroid') #show cluster centroid
plt.title('Clusters of Iris Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length(cm)')
plt.legend()
plt.show()
```
